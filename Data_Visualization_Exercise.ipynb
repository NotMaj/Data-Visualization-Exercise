```python
#PYTHON VISUALIZATIONS
df['Country'].value_counts()
```

```python
#Histogram for Mexico Females and their literacy rates
gender = df['Gender'] == 'female'
country = df['Country'] =='Mexico'
new_df = df.loc[gender & country]
```

```python
new_df
```

```python
sns.histplot(data = new_df, x='Literacy rate', color='lightblue', edgecolor ='black')
plt.title('Distribution of Literacy rates for Female Mexicans')
plt.axvline(df['Literacy rate'].median(),
             color='black',
             ls=':',
             label= f"Median is {round(df['Literacy rate'].median(),2)}")
plt.legend();
```

From the histogram, the female population of Mexico exhibits a constant high literacy rate as seen from the count of values closest to 1.

```python
df_2=df[['Year','Literacy rate']]
df_2
```

```python
grouped = df_2.groupby(['Year'])['Literacy rate'].nunique()
```

```python
grouped
```

```python
years = list(grouped.index)

years
```

```python
rates = list(grouped.values)
rates
```

```python
plt.style.use('seaborn')
plt.plot(years,rates, marker='o', markersize = 10)
plt.title('Reported literacy rates throughout the years')
plt.xlabel('DATES')
plt.ylabel('Reported Literacy rates');
```

From this graph, we can see how the data of reported literacy rates varies greatly, which can affect how the study of improvement or decline with literacy rates in certain countries.
